# **Labeling Rules**

*   **Noise**

    * Mengandung tawaran jual-beli, kupon, voucher, atau “cek link ini” dengan tujuan iklan atau promosi untuk mendorong transaksi.

    * Frasa seperti “klik di sini!”, “jangan lewatkan!”, “bagikan kalau setuju!” tanpa konteks diskusi.

    * Menggunakan hashtag trending tetapi isi tweet tidak berhubungan, hanya untuk menarik engagement.

    * Short URL (bit.ly, tinyurl, atau URL pendek lain) tanpa penjelasan isi link.

    * Menandai banyak akun (@user1 @user2 …) hanya untuk promosi.

    * Tweet hanya meneriakkan keyword tanpa membahas topik.
    
*   **Factual**

    * Melaporkan data atau observasi dengan kalimat deskriptif. Contohnya “#Grok menduduki peringkat #1 sejak pukul 20:00”.

    * Kalimat tanpa opini atau emosi murni statistik, fakta, atau informasi, serta tidak berisi pujian, kritik, atau keluhan.

    * Strukturnya menyerupai berita atau laporan resmi, biasanya menggunakan bahasa Formal atau semi-formal.

    * Tidak ada tawaran jual-beli, link komersial, atau ajakan aksi.

*   **Relevant**

    * Mengungkapkan suka, tidak suka, humor, atau pengalaman pribadi terkait topik.

    * Menyertakan alasan, kritik (positif atau negatif), atau rekomendasi terkait topik.

    * Berisi pertanyaan untuk diskusi, misalnya “Ada yang pernah ikut acara ini #KonserMalamMinggu? Bagaimana kesannya?”.

    * Berisi pengalaman pribadi atau observasi lapangan yang dikemas dalam bentuk cerita.

    * Pembahasan kontekstual yang membahas makna atau latar belakang topik trending.

    * Hashtag & mention digunakan relevan sesuai isi, bukan massal tanpa kaitan.

## **Emoji Extraction**

```python
import re
```

**Dataset Emoji**

```python
df_emoji = pd.read_csv("/content/drive/MyDrive/emoji_to_text.csv")
```

```python
emoji_map = dict(zip(df_emoji["emoji"], df_emoji["makna"]))
```

```python
emojis_sorted = sorted(emoji_map.keys(), key=len, reverse=True)
emoji_pattern = re.compile("|".join(map(re.escape, emojis_sorted)))
```

```python
def replace_emoji(text: str) -> str:
  return emoji_pattern.sub(lambda m: " " + emoji_map[m.group(0)] + " ", text)
```

## **Cleaning Data**

*   Mengganti entitas umum.
*   Mengganti karakter baris baru ke dalam spasi normal.
*   Menghapus semua token mulai dengan http:// atau https://.
*   Menghapus semua @username.
*   Menghapus tanda # tapi tetap pertahankan kata.
*   Menghapus “rt” di awal tweet kalau ada.
*   Menghapus karakter non-alphanumeric (kecuali spasi).
*   Menghilangkan angka jika tidak dibutuhkan.
*   Mengganti multi-spasi/tab/newline menjadi satu spasi.
*   Trim spasi di ujung/awal string.

```python
import html
```

```python
def clean_text(text: str) -> str:
  text = html.unescape(text)
  text = text.replace("\\n", " ").replace("\n", " ").replace("\r", " ")
  text = re.sub(r'http\S+', '', text)
  text = re.sub(r'@\w+', '', text)
  text = re.sub(r'#(\w+)', r'\1', text)
  text = re.sub(r'^\s*rt\s+', '', text)
  text = re.sub(r'[^\w\s]', ' ', text)
  text = re.sub(r'\d+', '', text)
  text = re.sub(r'\s+', ' ', text)

  return text.strip()
```

```python
df['text_cleaned'] = df['text_casefolded'].apply(clean_text)
```

```python
df['text_cleaned'] = df['text_cleaned'].fillna('').astype(str)
```

# **Threshold Tuning**

```python
from sklearn.metrics import f1_score
```

```python
probs        = lr_model.predict_proba(X_test)
classes      = lr_model.classes_
idx_factual  = np.where(classes == "factual")[0][0]
idx_relevant = np.where(classes == "relevant")[0][0]
```

```python
def predict_with_threshold(probs, tf, tr):
  preds = []

  for p in probs:
    if p[idx_factual] > tf:
      preds.append("factual")
    elif p[idx_relevant] > tr:
      preds.append("relevant")
    else:
      preds.append("noise")

  return np.array(preds)
```

## **Factual Threshold**

```python
best_tf, best_f1_fact = 0, 0
```

```python
for tf in np.linspace(0.1, 0.9, 17):
  is_fact_pred = probs[:, idx_factual] > tf
  is_fact_true = (y_test == "factual")
  f1_fact      = f1_score(is_fact_true, is_fact_pred)

  if f1_fact > best_f1_fact:
    best_f1_fact, best_tf = f1_fact, tf
```

```python
print(f"Best factual threshold: {best_tf:.2f} → factual F1={best_f1_fact:.3f}")
```

## **Relevant Threshold**

```python
best_tr, best_f1_rel = 0, 0
```

```python
for tr in np.linspace(0.1, 0.9, 17):
  is_rel_pred = (probs[:, idx_factual] <= best_tf) & (probs[:, idx_relevant] > tr)
  is_rel_true = (y_test == "relevant")
  f1_rel      = f1_score(is_rel_true, is_rel_pred)

  if f1_rel > best_f1_rel:
    best_f1_rel, best_tr = f1_rel, tr
```

```python
print(f"Best relevant threshold: {best_tr:.2f} → relevant F1={best_f1_rel:.3f}")
```

## **Predict with Optimal Threshold**

```python
y_pred_final = predict_with_threshold(probs, tf=best_tf, tr=best_tr)
```

```python
print(classification_report(y_test, y_pred_final, labels=classes))
print(confusion_matrix(y_test, y_pred_final))
```

```python
def stem_tokens(tokens: list[str]) -> list[str]:
  return [term_dict.get(t, stemmer.stem(t)) for t in tokens]
```

```python
def preprocess(raw: str) -> str:
  s      = normalize_unicode(raw)
  s      = replace_emoji(s)
  s      = case_fold(s)
  s      = clean_text(s)
  s      = normalize_slang(s)
  tokens = tweet_tokenizer.tokenize(s)
  tokens = [t for t in tokens if t]
  tokens = [t for t in tokens if len(t) > 2]
  tokens = remove_stopwords(tokens)
  tokens = stem_tokens(tokens)

  return " ".join(tokens)
```

```python
def predict_tweet(raw_text, artefact_path="/content/drive/MyDrive/twitter_trending_classifier.pkl"):
  obj        = joblib.load(artefact_path)
  tfidf      = obj["tfidf"]
  clf        = obj["clf"]
  tf         = obj["threshold_factual"]
  tr         = obj["threshold_relevant"]
  text_final = preprocess(raw_text)
  Xv         = tfidf.transform([text_final])
  probs      = clf.predict_proba(Xv)[0]
  classes    = clf.classes_
  p_fact     = probs[classes=="factual"][0]
  p_rel      = probs[classes=="relevant"][0]
  p_noise    = probs[classes=="noise"][0]

  if p_fact > tf:
    label = "factual"
  elif p_rel  > tr:
    label = "relevant"
  else:
    label = "noise"

  return {
      "label": label,
      "confidence": {
          "factual":  f"{p_fact*100:.1f}%",
          "relevant": f"{p_rel*100:.1f}%",
          "noise":    f"{p_noise*100:.1f}%"
      }
  }
```
